## Business Context 

The business model of Twitter is based on having an average of 330 million monthly active users using this platform to post messages called "tweets" of up to 280 characters and up to 2,400 times per day, which are automatically distributed to followers in a feed that is constantly refreshed, and then Twitter allows businesses to reach these users population who are then targeted as customers. 

Twitter divides its revenue into two categories: the sale of advertising services, which constitutes the vast majority of the company's revenue, as well as data licensing and other services.

The company creates tailored advertising opportunities by using an algorithm to make sure promoted products make it into the right users' timelines, "Who to Follow" lists, or at the top of the list of trending topics for an entire day in a particular country or globally. Advertisers also have the option of paying for in-stream video ads delivered to a targeted audience or sponsoring video content from publishing partners.

By analyzing the messages posted by users (tweets) businesses that are using Tweeter to advertize their products and services, can have almost instant feedback (negative, positive or neutral) about these products and services. However analyzing these tweets is a difficult task unless we apply tools such as sentiment analysis/classification that can help the brands to quickly detect negative social mentions that could harm their business.

Analysis of tweets with sentiment analysis/classification, which involves monitoring emotions in conversations on social media platforms, has become a key strategy in social media marketing.

Listening to how customers feel about the product/service on Twitter allows companies to understand their audience, keep on top of what’s being said about their brand and their competitors, and discover new trends in the industry.

## Sentiment Analysis Context


With the growth of social media, such as Twitter micro-blogs, millions of people connect and interact with each other anytime without geographical boundaries, as opposed to the traditional structured data available in databases. This resulted unstructured data requires new computational techniques for social media mining. Sentiment Analysis (a.k.a opinion mining) is a technique which automatically extracts and summarizes the opinions from very big volumes of data that the average human reader is unable to process.

Sentiment Analysis deals with analyzing people's sentiment, attitudes, emotions and opinions bout products, services, individuals, companies, organizations, events and topics and includes multiple fields such as Natural Language Processing (NLP), computational linguistics, information retrieval, machine learning and artificial intelligence.

The term "Sentiment Analysis" was coined by Nasukawa et al. in 2003. 

The esential issue in Sentiment Analysis is to identify how sentiments are expressed in texts and whether the expressions indicate positive or negative opinions about a particular subject, and this in term requires identification of sentiment expressions, polarity and strenght of expressions and their relationship to the subject.

In this project we are dealing with the sentiments expressed in tweets about problems of major U.S. airlines with tweets already classified as positive, negative and neutral by contributors.

As micro-blogs such as Twitter are limited to 280 characters they may not provide enough context to properly classify the text in case there are things such as irony, sarcasm, etc. These tweets have a lot of noise, which in term require good text cleaning procedures to make sure they are suitable for machine learning modeling.

```python
# Function to create seaborn label barplot displaying values in percentages

def mylabeled_barplot(data, feature):

    ax = sns.countplot(y=feature, data=data)
    plt.title('Bar Plot')
    plt.xlabel('Count')

    total = len(data[feature])
    for p in ax.patches:
            percentage = '{:.1f}%'.format(100 * p.get_width()/total)
            x = p.get_x() + p.get_width() + 0.02
            y = p.get_y() + p.get_height()/2
            ax.annotate(percentage, (x, y))

    plt.show()
```

```python
# Function to print unique values of a particular feature
def print_unique_values(data,feature):
    categorical_var = [feature,]
    for column in categorical_var:
        print(data[column].value_counts())
        print("-" * 40)
```

```python
# function to plot stacked bar chart


def stacked_barplot(data, predictor, target):
    """
    Print the category counts and plot a stacked bar chart

    data: dataframe
    predictor: independent variable
    target: target variable
    """
    count = data[predictor].nunique()
    sorter = data[target].value_counts().index[-1]
    tab1 = pd.crosstab(data[predictor], data[target], margins=True).sort_values(
        by=sorter, ascending=False
    )
    print(tab1)
    print("-" * 120)
    tab = pd.crosstab(data[predictor], data[target], normalize="index").sort_values(
        by=sorter, ascending=False
    )
    tab.plot(kind="bar", stacked=True, figsize=(count + 1, 5))
    plt.legend(
        loc="lower left",
        frameon=False,
    )
    plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
    plt.show()
```

# Vectorization

As raw text data cannot be fed directly to the machine learning algorithms, we need to convert them in numerical feature vectors with a fixed size rather than the raw text documents with variable lenght. Turning a collection of text documents into numerical feature vectors is called "vectorization".

In this project we apply **CountVectorizer** and **TfidfVectorizer** on the required text column to make it suitable for fitting the model. 

## Use CountVectorizer

The frequency of the word is counted here. If there are **d** unique words in our dictionary then for every review the vector will be of length **d** and count of word from reviews is stored at its particular location in vector. The vector will be highly sparse in such cases.

## Use TfidfVectorizer

Words like “And”, “For”, “The”, “a” etc. appears many times in a corpus and their large counts will not be very meaningful in the encoded vectors.

TF-IDF captures how important a word is to the document (without looking at other documents in the dataset).

TF-IDF (term frequency inverse document frequency) is a scheme to weight individual tokens.

One of the advantages of TF-IDF is to reduce the impact of tokens that occur very frequently, hence offering little to nothing in terms of information.


```python
# Let us inspect a sample text 
print('Sample text 4525: ')
print(df3.text[4525])
print()
print('Sample text 4525 length: ')
print(len(df3.text[4525]))
```

**The following function makes a plot of an sklearn Confusion Matrix cm using a Seaborn heatmap visualization**

```python
#This function makes a plot of an sklearn Confusion Matrix cm using a Seaborn heatmap visualization#
def make_confusion_matrix(cf,
                          group_names=None,
                          categories='auto',
                          count=True,
                          percent=True,
                          cbar=True,
                          xyticks=True,
                          xyplotlabels=True,
                          sum_stats=True,
                          figsize=None,
                          cmap='Blues',
                          title=None):

    # CODE TO GENERATE TEXT INSIDE EACH SQUARE
    blanks = ['' for i in range(cf.size)]

    if group_names and len(group_names)==cf.size:
        group_labels = ["{}\n".format(value) for value in group_names]
    else:
        group_labels = blanks

    if count:
        group_counts = ["{0:0.0f}\n".format(value) for value in cf.flatten()]
    else:
        group_counts = blanks

    if percent:
        group_percentages = ["{0:.2%}".format(value) for value in cf.flatten()/np.sum(cf)]
    else:
        group_percentages = blanks

    box_labels = [f"{v1}{v2}{v3}".strip() for v1, v2, v3 in zip(group_labels,group_counts,group_percentages)]
    box_labels = np.asarray(box_labels).reshape(cf.shape[0],cf.shape[1])


    # CODE TO GENERATE SUMMARY STATISTICS & TEXT FOR SUMMARY STATS
    if sum_stats:
        #Accuracy is sum of diagonal divided by total observations
        accuracy  = np.trace(cf) / float(np.sum(cf))

        #if it is a binary confusion matrix, show some more stats
        if len(cf)==2:
            #Metrics for Binary Confusion Matrices
            precision = cf[1,1] / sum(cf[:,1])
            recall    = cf[1,1] / sum(cf[1,:])
            f1_score  = 2*precision*recall / (precision + recall)
            stats_text = "\n\nAccuracy={:0.3f}\nPrecision={:0.3f}\nRecall={:0.3f}\nF1 Score={:0.3f}".format(
                accuracy,precision,recall,f1_score)
        else:
            stats_text = "\n\nAccuracy={:0.3f}".format(accuracy)
    else:
        stats_text = ""


    # SET FIGURE PARAMETERS ACCORDING TO OTHER ARGUMENTS
    if figsize==None:
        #Get default figure size if not set
        figsize = plt.rcParams.get('figure.figsize')

    if xyticks==False:
        #Do not show categories if xyticks is False
        categories=False


    # MAKE THE HEATMAP VISUALIZATION
#    plt.figure(figsize=figsize)
    plt.figure(figsize = (10,7))
    sns.heatmap(cf,annot=box_labels,fmt="",cmap=cmap,cbar=cbar,xticklabels=categories,yticklabels=categories)

    if xyplotlabels:
        plt.ylabel('True label')
        plt.xlabel('Predicted label' + stats_text)
    else:
        plt.xlabel(stats_text)
    
    if title:
        plt.title(title)
```

### Random Forest Classifier & Hyperparameter Tuning with CountVectorizer

Random Forest classifier model needs multiple hyperparameters like n_estimators, max_depth, min_samples_split etc which affect the model performance. We will tune the RF model based on ‘n_estimators’.  

To get a better measure of prediction accuracy (which we can use as a proxy for goodness of fit of the model), we can successively split the data in folds that we use for training and testing, this is called a KFold cross-validation.

We already imported the following libraries from sklearn.model_selection:
- KFold: Split dataset into k consecutive folds
- StratifiedKFold: Variation of Kfold that returns stratified folds
- cross_val_score: Evaluate a score by cross-validation


```python
# Finding optimal ‘n_estimators’ (NE) using k-fold cross-validated scores
# We start with an initial base of NE using np.arange & tolist
# np.arange returns evenly spaced values within a given interval
# tolist used to convert the data elements of an array into a list
base_ln = np.arange(100,400,100).tolist()
base_ln
```

```python
# K-Fold Cross - validation .
cv_scores = []
for b in base_ln:
    clf = RandomForestClassifier(n_estimators = b)
    scores = cross_val_score(clf, X1_train, y1_train, cv = 5, scoring = 'accuracy')
    cv_scores.append(scores.mean())
```

```python
# plotting the error as k increases
error = [1 - x for x in cv_scores]                                 #error corresponds to each number of n_estimator
optimal_learners = base_ln[error.index(min(error))]                #Selection of optimal nu of n_estimator corresponds to minimum error.
plt.plot(base_ln, error)                                           #Plot between each nu of estimator and misclassification error
xy = (optimal_learners, min(error))
plt.annotate('(%s, %s)' % xy, xy = xy, textcoords='data')
plt.xlabel("Number of base learners")
plt.ylabel("Misclassification Error")
plt.show()
```

```python
# Training the best model and calculating accuracy on test data .
rfclf1 = RandomForestClassifier(n_estimators = optimal_learners)
rfclf1.fit(X1_train, y1_train)
rfclf1.score(X1_test, y1_test)
count_vectorizer_predicted = rfclf1.predict(X1_test)
```
